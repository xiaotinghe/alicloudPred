首先，导入所需的包。

```python
import numpy as np
from mxnet import gluon, init, nd, metric, autograd, cpu
from mxnet.contrib import text
from mxnet.gluon import data as gdata, loss as gloss, nn, rnn, utils as gutils
import os
import pandas as pd
import random
import time
```

读取数据文件，并将所有NA用0填充。

```python
with open('dataset_campus_competition.txt', 'r') as f:
    raw_data = f.readlines()
raw_Data = []
for row in raw_data:
    cpu_performances = row.split()[1].replace('\"','').split(',')
    raw_Data.append([float(c) if c!= 'NA' else None for c in cpu_performances])
random.shuffle(raw_Data)
raw_Data = np.array(raw_Data)
df = pd.DataFrame(raw_Data)
df = df.fillna(0)
corpus_indices = np.array(df)
```

训练中用历史90小时数据预测接下来72小时数据，在数据允许的情况下，尽可能用更多的历史数据。

然后进行归一化操作。

```python
X = nd.concat(*[nd.array(corpus_indices[:,i-90:i]) for i in range(90, corpus_indices.shape[1] - 72)], dim=0)
x_mean, x_var = nd.moments(X, axes=1)
x_std = nd.sqrt(x_var)
X = (X - x_mean.expand_dims(1))/x_std.expand_dims(1)
Y = nd.concat(*[nd.array(corpus_indices[:,i:i+72]) for i in range(90, corpus_indices.shape[1] - 72)], dim=0)
Y = (Y - x_mean.expand_dims(1))/x_std.expand_dims(1)
```

切分训练集和测试集。

```python
train_data = gdata.ArrayDataset(X[len(X)//10:], Y[len(X)//10:])
test_data = gdata.ArrayDataset(X[:len(X)//10], Y[:len(X)//10])
train_iter = gluon.data.DataLoader(train_data, 64, shuffle=True, last_batch='rollover')
test_iter = gluon.data.DataLoader(test_data, 16, shuffle=False, last_batch='keep')
```

根据之前的经验，需求变化大概可以分为三种短期和长期不断重复的混合模式：

* 长期平稳变化的数据
* 短时间内即将到来的高峰
* 周期性变化的数据

为了适应这三种模式，采用时序预测的经典模型：LSTNet

Lai, Guokun, Chang, Wei-Cheng, Yang, Yiming, & Liu, Hanxiao. . Modeling long- and short-term temporal patterns with deep neural networks.

在本例中，LSTNet的GRU用来捕捉长期模式，SkipGRU用来捕捉周期性模式，AR用来捕捉短期模式。

由于需要预测未来72小时的，即这是个多步预测问题。在LSTNet基础上做了改进。主要是AR部分需要用到已经预测的结果（会带来误差累积的问题）

```python
class LSTNet(gluon.Block):
    def __init__(self, num_series, conv_hid, kernel_size, gru_hid, skip_gru_hid, skip, ar_window):
        super(LSTNet, self).__init__()
        kernel_size = kernel_size
        dropout_rate = 0.2
        self.skip = skip
        self.ar_window = ar_window
        with self.name_scope():
            self.conv = nn.Conv1D(conv_hid, kernel_size=kernel_size, layout='NCW', activation='relu')
            self.dropout = nn.Dropout(dropout_rate)
            self.gru = rnn.GRU(gru_hid, layout='TNC')
            self.skip_gru = rnn.GRU(skip_gru_hid, layout='TNC')
            self.fc = nn.Dense(72)
            self.ar_fc = nn.Dense(1)

    def forward(self, x):
        c = self.conv(x.transpose((0, 2, 1)))
        c = self.dropout(c)
        r = self.gru(c.transpose((2, 0, 1)))
        r = r[-1]  # Only keep the last output
        r = self.dropout(r)  # Now in NC layout
        # Skip GRU
        # Slice off multiples of skip from convolution output
        skip_c = c[:, :, -(c.shape[2] // self.skip) * self.skip:]
        skip_c = skip_c.reshape(c.shape[0], c.shape[1], -1, self.skip)  # Reshape to NCT x skip
        skip_c = skip_c.transpose((2, 0, 3, 1))  # Transpose to T x N x skip x C
        skip_c = skip_c.reshape(skip_c.shape[0], -1, skip_c.shape[3])  # Reshape to Tx (Nxskip) x C
        s = self.skip_gru(skip_c)
        s = s[-1]  # Only keep the last output (now in (Nxskip) x C layout)
        s = s.reshape(x.shape[0], -1)  # Now in N x (skipxC) layout
        # FC layer
        fc = self.fc(nd.concat(r, s))  # NC layout
        new_x = x
        res = []
        # Add autoregressive and fc outputs
        for i in range(fc.shape[1]):
            ar_x = new_x[:, -self.ar_window:, :]
            ar_x = ar_x.transpose((0, 2, 1))  # NCT layout
            ar_x = ar_x.reshape(-1, ar_x.shape[2])  # (NC) x T layout
            ar = self.ar_fc(ar_x)
            ar = ar.reshape(new_x.shape[0], -1)  # NC layout
            c_res = fc[:,i].expand_dims(1) + ar
            new_x = nd.concat(new_x, c_res.expand_dims(1), dim= 1)
            res.append(c_res)
        return nd.concat(*res, dim=1)
```

评估函数

```python
def evaluate(data_iter):
    mean_absolute_error = metric.MAE()
    for i, (features, labels) in enumerate(data_iter):
        features = features.as_in_context(ctx)
        labels = labels.as_in_context(ctx)
        pred = net(features.expand_dims(-1))
        mean_absolute_error.update(labels = labels.reshape(-1), preds = pred.reshape(-1))
    return mean_absolute_error.get()[1]
```

开始训练，如果没有GPU，请修改第一行的gpu()为cpu()

```python
ctx = cpu()
net = LSTNet(num_series=1, conv_hid=16, kernel_size = 6, gru_hid=4, skip_gru_hid=16, skip=16, ar_window=24)
net.initialize(ctx=ctx)
loss = gloss.L1Loss()
trainer = gluon.Trainer(net.collect_params(), 'adam', {'learning_rate': 0.01})

num_batches, timer = len(train_iter), d2l.Timer()
num_epochs = 30
mean_absolute_error = metric.MAE()
best_mae = float('inf')
for epoch in range(num_epochs):
    # Store training_loss, training_accuracy, num_examples, num_features
    metric_temp = d2l.Accumulator(2)
    mean_absolute_error.reset()
    for i, (features, labels) in enumerate(train_iter):
        timer.start()
        features = features.as_in_context(ctx)
        labels = labels.as_in_context(ctx)
        with autograd.record():
            pred = net(features.expand_dims(-1))
            l = loss(pred, labels)
        l.backward()
        trainer.step(labels.shape[0], ignore_stale_grad=True)

        metric_temp.add(l.sum().asscalar(), labels.shape[0])
        mean_absolute_error.update(labels = labels.reshape(-1), preds = pred.reshape(-1))
        timer.stop()
    test_mae = evaluate(test_iter)
    if test_mae < best_mae:
        best_mae = test_mae
        #保存模型参数
        net.save_parameters('net.params')
print('loss %.3f, train acc %.3f, test acc %.3f' % (
    metric_temp[0] / metric_temp[1], mean_absolute_error.get()[1], test_mae))
```
